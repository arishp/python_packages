```python
data = {
    'A': [1, 2, 3],
    'B': [4, 5, 6],
    'C': [7, 8, 9]}

# Convert the dictionary into DataFrame
df = pd.DataFrame(data)
print("Original DataFrame:\n", df)
```

```python
# Function to add
def add_values(row):
	return row['A'] + row['B'] + row['C']
```

```python
# Apply the user-defined function to every row
df['add'] = df.apply(add_values, axis=1)
```

```python
print('\nAfter Applying Function: ')
# Print the new DataFrame
print(df)
```

```python
import matplotlib.pyplot as plt 
x = [3, 1, 3] 
y = [3, 2, 1] 

plt.plot(x, y)
plt.title("Line Chart")

plt.legend(["Line"])
plt.show()
```

```python
import matplotlib.pyplot as plt 
x = [3, 1, 3, 12, 2, 4, 4] 
y = [3, 2, 1, 4, 5, 6, 7] 

plt.bar(x, y)

plt.title("Bar Chart")

plt.legend(["bar"])
plt.show()
```

```python
import matplotlib.pyplot as plt 
x = [1, 2, 3, 4, 5, 6, 7, 4] 

plt.hist(x, bins = [1, 2, 3, 4, 5, 6, 7])

plt.title("Histogram")
plt.legend(["bar"])

plt.show()
```

```python
import matplotlib.pyplot as plt 

x = [3, 1, 3, 12, 2, 4, 4]
y = [3, 2, 1, 4, 5, 6, 7]
plt.scatter(x, y)

plt.legend("A")

plt.title("Scatter chart")
plt.show()
```

```python
import matplotlib.pyplot as plt 
x = [1, 2, 3, 4] 

e  =(0.1, 0, 0, 0)

plt.pie(x, explode = e)

plt.title("Pie chart")
plt.show()
```
